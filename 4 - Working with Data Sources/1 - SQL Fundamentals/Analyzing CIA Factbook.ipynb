```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("factbook.db")
query = "SELECT * FROM sqlite_master WHERE type='table';"
pd.read_sql_query(query, conn)
```

```python
facts_rows_query = "SELECT * FROM facts LIMIT 5"
pd.read_sql_query(facts_rows_query, conn)
```

```python
query3 = "SELECT MIN(population) Min_Population, MAX(population) Max_Population, MIN(population_growth) Min_PopGrowth, MAX(population_growth) Max_PopGrowth FROM facts"
pd.read_sql_query(query3, conn)
```

```python
##Write Query for country with minimum population
query4 = "SELECT name, population FROM facts WHERE population = (SELECT Min(Population) FROM facts)"
pd.read_sql_query(query4, conn)
```

```python
##Write Query for country with maximum population
query5 = "SELECT name, population FROM facts WHERE population = (SELECT Max(Population) FROM facts)"
pd.read_sql_query(query5, conn)
```

```python
#Calculation average population and area
query6 = "SELECT AVG(population) AVG_Population, AVG(area) AVG_Area FROM facts"
pd.read_sql_query(query6, conn)
```

```python
##Select all countries where population is greater than average, and are is below average"
query7 = "SELECT name, population, area FROM facts WHERE population > (SELECT AVG(population) FROM facts) AND area < (SELECT AVG(area) FROM facts)"
pd.read_sql_query(query7, conn)
```
